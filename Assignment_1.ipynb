### Python Part 1

````python
import numpy as np
np.random.seed(0)  # seed for reproducibility

for i in range(5):
    a1 = np.random.randint(10, size=5)  # One-dimensional array. For representing the numbers in the die throw five times
    a2 = np.random.randint(10, size=5) 
    C = a1 - a2

    print ("Player 1", "\n", a1) #a1 = Player 1
    print ("Player 2", "\n", a2) #a2 = Player 2

    counta1 = np.count_nonzero(C > 0) #counts the number of occurences
    counta2 = np.count_nonzero(C < 0)

    if counta1==counta2:
        print("It is a draw with", (counta1), "game each", "\n") #compares the results of the two players' die throws per game
    
    elif counta1>counta2:
        print("Player 1 wins by", (counta1 - counta2), "game", "\n")
        
    else:
        print("Player 2 wins by", (counta2 - counta1), "game", "\n")

````
